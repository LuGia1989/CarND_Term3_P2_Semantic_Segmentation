## Data Exploration after Semantic Segmentation

```python
# Importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import random
```

```python
images =[]
for imagePath in glob.glob('./runs/1510499693.5846434' + '/*.png'):
    image = mpimg.imread(imagePath)
    images.append(image)

for i in range(8):
    j = (random.randint(1,289))
    plt.subplot(4, 2, i+1)
    plt.imshow(images[j],interpolation='none', aspect='auto')
plt.show()
```

```python
# More photo
for i in range(6):
    j = (random.randint(1,289))
    plt.subplot(3, 2, i+1)
    plt.imshow(images[j],interpolation='nearest', aspect='auto')
plt.show()
```
